```python
model1 = RandomForestRegressor()
model2 = RandomForestRegressor()

param_grid1 = [
             {"n_estimators": [10,80], #add the equal dirtstubutiun to this, check ipad notes 
                'max_features': ["sqrt"],
                'max_depth': [2,20],
                'max_leaf_nodes': [2,60],
                "min_samples_split": [2,8],
                "min_samples_leaf":[1,40]
             }
]
```

```python
search1 = RandomizedSearchCV(model1,param_grid1, n_iter = 7, cv=9, scoring = "r2")
search1.fit(X_train,y_train)
 
```

```python
hyperameters1 = search1.best_params_
```

```python
plt.figure()
plt.scatter(y_train, y_score_train1, alpha = 0.5)
min_val = min(y_train.min(), y_score_train1.min())
max_val = max(y_train.max(), y_score_train1.max())
plt.plot([min_val, max_val], [min_val, max_val], 'k--', lw=2, label='y = x')
plt.title('R^2 for True vs Predicted Values for Model 1')
# plt.legend(plt.legend(['Line of Best Fit', 'R^2 = {:.2f}'.format(r2)]))
plt.legend(['R^2 Train = {:0.2f}'.format(r2_score(y_train, y_score_train1)),"Y=X Treadline"])
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
```

```python
def perform_cross_validation(model, X, y, scoring, cv=5):
    scores = cross_val_score(model, X, y, scoring=scoring, cv=cv)
    return scores
```

```python
scoring = make_scorer(r2_score)

scores1 = perform_cross_validation(model1, X, y, scoring=scoring, cv=5)
scores2 = perform_cross_validation(model2, X, y, scoring=scoring, cv=5)
```

```python
cv_results = pd.DataFrame({
    'Model 1': scores1,
    'Model 2': scores2
})

cv_results.plot(kind='bar', figsize=(10, 6))
plt.title('Cross-Validation R² Scores')
plt.xlabel('Fold')
plt.ylabel('R² Score')
plt.legend(title='Model')
plt.show()
```
